# RPM 0% Noise

```python
from scipy.io import loadmat, savemat
import os
import numpy as np
import math
import random
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
from torch.distributions.multivariate_normal import MultivariateNormal
from torch.distributions.normal import Normal
from torch.distributions.kl import kl_divergence
from torch.nn import BCEWithLogitsLoss
from util import VNet3
from torch.distributions.beta import Beta
import pandas as pd

def evaluate_likelihood(Y, score):#, sigma
    s = torch.sigmoid(score)
    likelihood = s ** Y * (1 - s) ** (1 - Y)
    return likelihood

test_acc = []

for k in range(5):
    
    
    data = loadmat('./cancer_{}.mat'.format(k+1))
    X_train = torch.tensor(data['X_train'], dtype=torch.float32)
    y_train = torch.tensor(data['y_train'], dtype=torch.float32)
    X_meta = torch.tensor(data['X_meta'], dtype=torch.float32)
    y_meta = torch.tensor(data['y_meta'], dtype=torch.float32)
    X_test = torch.tensor(data['X_test'], dtype=torch.float32)
    y_test = torch.tensor(data['y_test'], dtype=torch.float32)
    print(X_train.shape[0], X_meta.shape[0], X_test.shape[0])
    
    l = 1000

    Y_train = y_train.repeat(l, 1).t()
    Y_test = y_test.repeat(l, 1).t()
    Y_meta = y_meta.repeat(l, 1).t()
    
    n, p = np.shape(X_train)
    p = p + 1
    
    
    lam = 100
    beta_prior = MultivariateNormal(loc=torch.zeros(p), covariance_matrix=lam * torch.eye(p))
    v_prior = Beta(0.01 * torch.ones(n), 0.01 * torch.ones(n))

    
    mu = torch.zeros(p, requires_grad=True)
    v_log_diag = torch.rand([p], requires_grad=True)
    v_non_diag = torch.randn([int(p * (p - 1) / 2)], requires_grad=True)
    L_indices_non_diag = torch.tril_indices(row=p, col=p, offset=-1)
    a = torch.rand(n, requires_grad=True)
    b = torch.rand(n, requires_grad=True)

    
    likelihood = BCEWithLogitsLoss(reduction='sum')

    
    l = 1000
    Y_train = y_train.repeat(l, 1).t()

    optimizer = torch.optim.Adam([mu, v_log_diag, v_non_diag, a, b], lr=0.01)

    num_iters = 1000

    loss_all = []

    for iters in range(num_iters):
        
        L = torch.diag(torch.exp(v_log_diag))
        L[L_indices_non_diag[0], L_indices_non_diag[1]] = v_non_diag

        
        beta_post = MultivariateNormal(loc=mu, scale_tril=L)

        
        beta = beta_post.rsample(sample_shape=[l])
    
        v_post = Beta(a, b)
        v = v_post.rsample(sample_shape=[l])

        # pred_logit = torch.matmul(X_train, beta.t())
        pred_logit = torch.matmul(X_train, beta[:,:p-1].t()) + beta[:,p-1].repeat(X_train.shape[0], 1)
        likeli = evaluate_likelihood(Y_train, pred_logit) + 1e-40
        loss = - (v.mean(dim = 0) * torch.log(likeli).mean(dim = 1)).mean() + kl_divergence(beta_post, beta_prior) + kl_divergence(v_post, v_prior).sum()
    
        print(iters, loss.item(), mu[:2])

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        loss_all.append(loss.item())

    print(loss_all[0])
    print(loss_all[-1])
    print(mu)

    plt.plot(loss_all)
    plt.show()
    
    y_test[y_test == 0] = -1
    
    def acc_cal(a,b):
        
        l = a.shape[0]
        assert l == b.shape[0]
    
        t = torch.sign(a * b)
        hit = t[t>0].shape[0]
        # hit = torch.sign(a*b).sum()
    
        return hit / l

    beta_post = MultivariateNormal(loc = mu, scale_tril = L)
    wb = beta_post.rsample(sample_shape=[l])
    aa = torch.matmul(X_test, wb[:,:p-1].t()) + wb[:,p-1].repeat(X_test.shape[0], 1)
    acc = acc_cal(aa.mean(dim = 1), y_test[0])
    test_acc.append(acc)

    print(k+1, test_acc)
```

```python
print(test_acc)
test_acc = np.array(test_acc)
print('{}({})'.format(test_acc.mean(), test_acc.std()))
```
